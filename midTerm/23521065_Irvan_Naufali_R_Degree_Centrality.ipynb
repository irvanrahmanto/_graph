```python
# Import library
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt

# Import data yang digunakan
# Data yang digunakan merupakan facebook combined dengan format .txt
dataRead = pd.read_csv('facebook_combined.txt', sep="  ")
dataRead.head()
```

```python
G = nx.read_edgelist('facebook_combined.txt', create_using=nx.Graph())
nx.draw_networkx(G)
```

```python
plt.figure(figsize=(25, 25))
pos=nx.spring_layout(G, k=0.15)
nx.draw_networkx(G,pos,node_size=25, node_color='blue')
plt.show()
```

```python
deg_cent = nx.degree_centrality(G)

plt.figure(figsize=(30,30))
plt.ylabel("Count", fontsize = 20)
plt.xlabel("Degree", fontsize=20)
plt.hist(list(deg_cent.values()))

plt.title("Degree Centrality Historgam", fontsize=35)
plt.show()
```

```python
def plot_degree_dist(G):
  degrees = [G.degree(n) for n in G.nodes()]
  plt.figure()
  plt.hist(degrees)
  plt.show()

plot_degree_dist(nx.gnp_random_graph(100, 0.5, directed=True))
```
