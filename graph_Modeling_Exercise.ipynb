```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import File .csv
rfd = pd.read_csv('data.csv')
my_graph = nx.from_pandas_edgelist(rfd, source="Overall", target="Age")
pos = nx.bipartite_layout(my_graph, rfd.Overall)
nx.draw(my_graph, pos=pos, with_labels=True)
```
